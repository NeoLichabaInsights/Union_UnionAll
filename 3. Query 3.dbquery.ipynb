```sql
SELECT *
FROM union_unionall.banking.retail_banking_staff;

SELECT *
FROM union_unionall.banking.corporate_banking_staff;

--Create a unique list of all staff who work in either retail banking or corporate banking. 

SELECT staff_id,
       staff_name,
       LOWER(REPLACE(staff_name, " ",  "." )|| "@banking.com") AS email
FROM union_unionall.banking.retail_banking_staff

UNION 
SELECT staff_id,
       staff_name,
       LOWER(REPLACE(staff_name, " ",  ".") || "@banking.com") AS email
FROM union_unionall.banking.corporate_banking_staff;
```
