```sql
SELECT *
FROM union_unionall.banking.branch_rosebank_accounts;

SELECT *
FROM union_unionall.banking.branch_sandton_accounts;

-- Combine account holders from both branches into one unique list. 

SELECT account_id, 
       account_holder,
       'Rosebank' AS city
FROM union_unionall.banking.branch_rosebank_accounts
UNION
SELECT account_id, 
       account_holder,
       'Sandton' AS city
FROM union_unionall.banking.branch_sandton_accounts;
```
