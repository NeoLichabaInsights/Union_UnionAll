```sql
SELECT
  *
FROM
  union_unionall.banking.may_payments;

SELECT
  *
FROM
  union_unionall.banking.april_payments;

-- Reconcile all payment records from April and May into one table for the half-year report.
-- The total count row matters for reconciliation
SELECT
  payment_id,
  account_id,
  amount,
  "April" AS payment_date
FROM
  union_unionall.banking.april_payments
UNION ALL
SELECT
  payment_id,
  account_id,
  amount,
  "May" AS payment_date
FROM
  union_unionall.banking.may_payments;

-- Total number of payments
SELECT
  (
    SELECT
      COUNT(*)
    FROM
      union_unionall.banking.april_payments
  )
    + (
      SELECT
        COUNT(*)
      FROM
        union_unionall.banking.may_payments
    ) AS total_payments;
```
