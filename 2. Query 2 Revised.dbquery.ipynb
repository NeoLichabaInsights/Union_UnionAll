```sql
-- Run code to check contents of tables
SELECT
  *
FROM
  union_unionall.banking.savings_products;

SELECT
  *
FROM
  union_unionall.banking.current_products;

-- Build a unique catalogue that shows that every product the bank offers without repeating any product that appears in both lists
SELECT
  product_code,
  product_name,
  CASE
    WHEN product_name LIKE '%Savings%' THEN 'Savings'
    WHEN product_name LIKE '%Current%' THEN 'Current'
    ELSE product_name
  END AS product_type
FROM
  union_unionall.banking.savings_products
UNION
SELECT
  product_code,
  product_name,
  CASE
    WHEN product_name LIKE '%Savings%' THEN 'Savings'
    WHEN product_name LIKE '%Current%' THEN 'Current'
    ELSE product_name
  END AS product_type
FROM
  union_unionall.banking.current_products;
```
