```sql
SELECT
  *
FROM
  union_unionall.banking.app_complaints;

SELECT
  *
FROM
  union_unionall.banking.email_complaints;

-- Compile a full record of every complaint logged this month, whether via email or via mobile app.
-- The team needs all complaints to calculate total complaint volumes
SELECT
  complaint_id,
  customer_id,
  "Email Compliants" AS category,
  NULL as logged_date
FROM
  union_unionall.banking.email_complaints
UNION ALL
SELECT
  complaint_id,
  customer_id,
  "App Compliants" AS category,
  NULL as logged_date
FROM
  union_unionall.banking.app_complaints;

-- Total number of complaints
SELECT
  (
    SELECT
      COUNT(*)
    FROM
      union_unionall.banking.email_complaints
  )
    + (
      SELECT
        COUNT(*)
      FROM
        union_unionall.banking.app_complaints
    ) AS total_complaints;
```
