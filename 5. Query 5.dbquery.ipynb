```sql
SELECT *
FROM union_unionall.banking.push_notification_targets;

SELECT *
FROM union_unionall.banking.inapp_banner_targets;

-- Build a unique customer list of every customer targeted via push notification and/or in-app banner.

SELECT customer_id, customer_name,
       'Push Notification' AS segment
FROM union_unionall.banking.push_notification_targets

UNION

SELECT customer_id, customer_name,
    'In-app Banner' AS segment
FROM union_unionall.banking.inapp_banner_targets;
```
