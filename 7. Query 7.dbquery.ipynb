```sql
SELECT
  *
FROM
  union_unionall.banking.gauteng_loan_applications;

SELECT
  *
FROM
  union_unionall.banking.western_cape_loan_applications;

-- Create a list of all loan applicatios received from the Gauteng and Western Cape region.
-- Duplicates are welcomed
SELECT
  app_id AS application_id,
  customer_id,
  loan_type,
  NULL as amount_requested,
  "Gauteng" AS region
FROM
  union_unionall.banking.gauteng_loan_applications
UNION ALL
SELECT
  app_id AS application_id,
  customer_id,
  loan_type,
  NULL as amount_requested,
  "Western Cape" AS region
FROM
  union_unionall.banking.western_cape_loan_applications;
```
