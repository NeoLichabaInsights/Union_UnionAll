```sql
SELECT
  *
FROM
  union_unionall.banking.mobile_branch_cities;

SELECT
  *
FROM
  union_unionall.banking.digital_branch_cities;

-- Provide a unique list of all cities where the bank serves customers - either through its mobile or digital branches
SELECT
  city_code,
  city_name,
  city_name AS region
FROM
  union_unionall.banking.mobile_branch_cities
UNION
SELECT
  city_code,
  city_name,
  city_name AS region
FROM
  union_unionall.banking.digital_branch_cities;
```
